```python
import pandas as pd

# membaca dataset dan mengubahnya menjadi dataframe
data = pd.read_csv('/content/drive/MyDrive/Dataset/Social_Network_Ads.csv')
```

```python
#melihat lima data teratas
data.head()
```

```python
#mengecek nilai kosong pada setiap atribut
data.info()
```

```python
#hapus kolom yang tidak penting untuk dipelajari model
data = data.drop(columns=['User ID'])
```

```python
#one hot encoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
# pisahkan atribut dan label
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']
X = data[predictions]
y = data['Purchased']
```

```python
# lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns = X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

#bagi data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
#membuat model logistic regression

from sklearn import linear_model

#latih dengan fungsi fit()
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
#uji akurasi model
model.score(X_test, y_test)
```
